```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
```

```python
Fs = 1000 # sampling frequency [Hz]
f1 = 50 # signal 1 frequency [Hz]
f2 = 3*f1 # signal 2 frequency [Hz]
f3 = 9*f1 # signal 3 frequency [Hz]
t_max = 2 # measurement time [s]
A = 2 # amplitude [m/s]


k = 1/Fs
t = np.arange(0, t_max, k)
g = np.arange(0.999, 3-k, k)
p = 2
omega1 = 2*np.pi*f1*g/p
omega2 = 2*np.pi*f2*g/p
omega3 = 2*np.pi*f3*g/p

x = A*np.sin(omega1*t) + A*np.sin(omega2*t) + A*np.sin(omega3*t)

plt.plot(t, x)
plt.title('Signal waveform')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()


n = len(x)
y = abs(fft(x))/(n/2)
y = y[0:int(n/2)]

plt.plot(y)
plt.title('Fourier transform')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Normalized amplitude')
plt.show()


plt.specgram(x, Fs = Fs);
plt.xlim(0.1, 1.8)
plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')
plt.show()
```
